```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from tqdm import tqdm
```

```python
def getInfo(url):
  page = requests.get(url, headers = {'User-Agent': 'My User Agent 1.0'})
  soup = BeautifulSoup(page.text, 'html')
  try:
    name = soup.find_all("h1", {"class":"tdb-title-text"})[0].text
  except:
    name = url.split("/")[-2]
  paragraph = soup.find_all("p")[0].text

  temp = soup.find_all("li", {"class": "wprm-recipe-ingredient"})
  ingredients = ["- " + i.text + "\n" for i in temp]

  temp = soup.find_all("li", {"class": "wprm-recipe-instruction"})
  instruction = [str(i+1) + ". " + temp[i].text +"\n" for i in range(len(temp))]

  temp = soup.find_all("span", {"class": "wprm-recipe-keyword wprm-block-text-normal"})
  try:
    keyword = temp[0].text
  except:
    keyword = ""
  # print(paragraph)
  # print(ingredients)
  # print(instruction)
  # print(keyword)
  output = {
    "Name": name,
    "URL": url,
    "Description": paragraph,
    "Ingredients": "".join(ingredients),
    "Instructions": "".join(instruction),
    "Keywords": keyword,
  }
  return output
```

```python
urls = pd.read_csv('file1.csv')
df = pd.DataFrame( columns=["Name","URL","Description","Ingredients","Instructions","Keywords"])
#print(df)
for i in urls["URL"]:
  df = df.append(getInfo(i), ignore_index = True)

df
df.to_csv('data.csv')
```
